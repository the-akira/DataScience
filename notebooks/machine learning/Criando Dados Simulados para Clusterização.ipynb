# Criando Dados Simulados para Clusterização

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

### Criando Dados

```python
# Criando as características (X) e o output (y) com 200 amostras
X, y = make_blobs(n_samples = 200,
                  # duas variáveis de característica
                  n_features = 2,
                  # três clusters
                  centers = 3,
                  # com .5 de desvio padrão de cluster
                  cluster_std = 0.5,
                  # shuffled,
                  shuffle = True)
```

```python
# Criando um gráfico de dispersão da primeira e segunda características
plt.scatter(X[:,0],
            X[:,1])

# mostra o gráfico de dispersão
plt.show()
```
