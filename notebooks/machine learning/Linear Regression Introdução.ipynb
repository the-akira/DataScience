# Linear Regression Introdução

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
data = pd.read_csv('https://dadosdatascience.netlify.com/cost_revenue_clean.csv')
```

```python
data.head(10)
```

```python
data.describe()
```

```python
X = pd.DataFrame(data, columns=['production_budget_usd'])
y = pd.DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.5)
plt.title('Film Cost vs GLobal Revenue')
plt.xlabel('Production Budget $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.ylabel('Worldwide Gross $')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
regression = LinearRegression()
```

```python
regression.fit(X, y)
```

Slope Coefficient:

```python
regression.coef_[0][0] # theta_1
```

Intercept:

```python
regression.intercept_[0]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.5, color='g')
plt.plot(X, regression.predict(X), color='k', linewidth=4)
plt.title('Film Cost vs GLobal Revenue')
plt.xlabel('Production Budget $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.ylabel('Worldwide Gross $')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
regression.score(X, y)
```
