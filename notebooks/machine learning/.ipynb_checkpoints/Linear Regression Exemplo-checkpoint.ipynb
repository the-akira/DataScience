```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from pylab import rcParams
%matplotlib inline
rcParams['figure.figsize'] = 10, 7

preco_casa = [245, 312, 279, 308, 199, 219, 405, 324, 319, 255]
tamanho = [1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]

print(tamanho)
```

```python
tamanho2 = np.array(tamanho).reshape((-1,1))
print(tamanho2)
```

```python
regr = linear_model.LinearRegression()
regr.fit(tamanho2, preco_casa)
print("Coeficientes: \n", regr.coef_)
print("Intercept: \n", regr.intercept_)
```

```python
novo_tamanho = 1400
preco = (novo_tamanho * regr.coef_) + regr.intercept_
print(preco)
```

```python
print(regr.predict([[novo_tamanho]]))
```

```python
# Fórmula obtida para o modelo treinado
def graph(formula, x_range):
    x = np.array(x_range)
    y = eval(formula)
    plt.plot(x, y)

# Plotando a linha de previsão
graph('regr.coef_*x + regr.intercept_', range(1000, 2700))
plt.scatter(tamanho, preco_casa, color='black')
plt.ylabel('preço da casa')
plt.xlabel('tamanho da casa')
plt.show()
```
