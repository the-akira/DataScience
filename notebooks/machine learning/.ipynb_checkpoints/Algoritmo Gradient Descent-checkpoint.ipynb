# O Algoritmo Gradient Descent

```python
# Imports necessários
import matplotlib.pyplot as plt 
import numpy as np
from sympy import diff

%matplotlib inline
```

### Exemplo 1 - Função de custo

#### $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

#### Gerando Dados

```python
x_1 = np.linspace(start=-3, stop=3, num=100)
print(x_1)
```

```python
plt.figure(figsize=(11,7))
plt.style.use('classic')
plt.grid()

plt.xlim([-3, 3])
plt.ylim(0, 8)

plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=2)
```

### Inclinação e Derivada

Vamos definir uma função derivada de $f(x)$ chamada $df(x)$

```python
# Calculando a derivada da função f(x) com o método diff da biblioteca SymPy
x = Symbol('x')
diff(x**2 + x + 1, x)
```

#### $$df(x) = 2x + 1$$

```python
# Definindo a função derivada
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=(11,7))
plt.style.use('classic')
plt.grid()
plt.title('Função Derivada')

plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)

plt.plot(x_1, df(x_1), color='blue', linewidth=2)
```

```python
plt.figure(figsize=[18,6])

# Função de Custo
plt.subplot(1, 2, 1)
plt.plot(x_1, f(x_1), c='g', linewidth=2)
plt.title('Função de Custo')
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid()

# Função derivada
plt.subplot(1, 2, 2)
plt.title('Função derivada')
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1, df(x_1), c='r', linewidth=2)
plt.grid()
```

```python
plt.figure(figsize=[8,18])

# Função de Custo
plt.subplot(2, 1, 1)
plt.plot(x_1, f(x_1), c='g', linewidth=2)

plt.title('Função de Custo')
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid()

# Função derivada
plt.subplot(2, 1, 2)
plt.title('Função derivada')
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1, df(x_1), c='r', linewidth=2)
plt.grid()
```

### Gradient Descent

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x 
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient    
    step_size = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print(f'Código rodou {n} vezes')
        break

print(f'O Local minimum ocorre em: {new_x}')
print(f'O valor da Inclinação ou df(x) nesse ponto é: {df(new_x)}')
print(f'O valor de f(x) ou função de custo nesse ponto é: {f(new_x)}')
```

```python
plt.figure(figsize=[20,6])

# Função de Custo
plt.subplot(1, 3, 1)
plt.plot(x_1, f(x_1), c='g', linewidth=2, alpha=0.8)
values = np.array(x_list)
plt.scatter(x_list, f(values), color='k', linewidth=4)

plt.title('Função de Custo')
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid()

# Função derivada
plt.subplot(1, 3, 2)
plt.title('Função derivada')
plt.xlim([-3, 3])
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1, df(x_1), c='r', linewidth=2, alpha=0.8)
plt.scatter(x_list, slope_list, color='k', linewidth=4)
plt.grid()

# Aproximação
plt.subplot(1, 3, 3)
plt.title('Gradient Descent (Aproximação)', fontsize=20)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)
plt.plot(x_1, df(x_1), color='black', linewidth=7, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.7)
```

### Exemplo 2 - Multiple Minima vs Palpite Inicial e Funções Avançadas

#### $$g(x) = x^4 - 4x^2 + 5$$

#### Gerando Dados

```python
x_2 = np.linspace(-2, 2, 1000)
```

```python
# Computando a derivada da função
x = Symbol('x')
diff(x**4 -4*x**2 + 5, x)
```

```python
# Definindo g(x)
def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
plt.figure(figsize=[18,6])

# Função de Custo
plt.subplot(1, 2, 1)
plt.plot(x_2, g(x_2), c='g', linewidth=2)

plt.title('Função de Custo')
plt.xlim([-2, 2])
plt.ylim(0.5, 5.5)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.grid()

# Função derivada
plt.subplot(1, 2, 2)
plt.title('Função derivada')
plt.xlim([-2, 2])
plt.ylim(-6, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.plot(x_2, dg(x_2), c='r', linewidth=2)
plt.grid()
```

### Gradient Descent como uma Função Python

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iter=300): 
    """Função Gradient Descent"""
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x 
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient    
        step_size = abs(new_x - previous_x)

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, 0.5, 0.02, 0.001)
print(f'Local Min ocorre em: {local_min}')
print(f'Número de passos: {len(list_x)}')
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=-0.5, multiplier=0.02, precision=0.001)
print(f'Local Min ocorre em: {local_min}')
print(f'Número de passos: {len(list_x)}')
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=-0.1)
print(f'Local Min ocorre em: {local_min}')
print(f'Número de passos: {len(list_x)}')
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=0.1)

plt.figure(figsize=[18,6])

# Função de Custo
plt.subplot(1, 2, 1)
plt.plot(x_2, g(x_2), c='g', linewidth=2, alpha=0.9)
plt.scatter(list_x, g(np.array(list_x)), color='red', s=50, alpha=0.7)

plt.title('Função de Custo')
plt.xlim([-2, 2])
plt.ylim(0.5, 5.5)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.grid()

# Função derivada
plt.subplot(1, 2, 2)
plt.title('Função derivada')
plt.xlim([-2, 2])
plt.ylim(-6, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.plot(x_2, dg(x_2), c='b', linewidth=2, alpha=0.7)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)
plt.grid()
```

### Exemplo 3 - Divergência e Overflow

#### $$h(x) = x^5 -2x^4 + 2$$

#### Gerando os Dados

```python
x_3 = np.linspace(start=-2.5, stop=2.5, num=1000)

def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5*x**4 - 8*x**3
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dh, initial_guess=-0.2, max_iter=70)

plt.figure(figsize=[18,6])

# Função de Custo
plt.subplot(1, 2, 1)
plt.plot(x_3, h(x_3), c='g', linewidth=2, alpha=0.9)
plt.scatter(list_x, h(np.array(list_x)), color='red', s=50, alpha=0.7)

plt.title('Função de Custo')
plt.xlim([-1.2, 2.5])
plt.ylim(-1, 4)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.grid()

# Função derivada
plt.subplot(1, 2, 2)
plt.title('Função derivada')
plt.xlim([-1, 2])
plt.ylim(-4, 5)
plt.xlabel('X', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.plot(x_3, dh(x_3), c='b', linewidth=2, alpha=0.7)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)
plt.grid()

print(f'Local min ocorre em {local_min}')
print(f'Custo nesse mínimo é:  {h(local_min)}')
print(f'Número de passos: {len(list_x)}')
```

### A Taxa de Aprendizado

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=1.9, multiplier=0.02, max_iter=500)

plt.figure(figsize=[18,6])

# Função de Custo
plt.subplot(1, 2, 1)
plt.plot(x_2, g(x_2), c='g', linewidth=2, alpha=0.9)
plt.scatter(list_x, g(np.array(list_x)), color='red', s=50, alpha=0.7)

plt.title('Função de Custo')
plt.xlim([-2, 2])
plt.ylim(0.5, 5.5)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.grid()

# Função derivada
plt.subplot(1, 2, 2)
plt.title('Função derivada')
plt.xlim([-2, 2])
plt.ylim(-6, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.plot(x_2, dg(x_2), c='b', linewidth=2, alpha=0.7)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)
plt.grid()

print(f'Número de passos é: {len(list_x)}')
```

```python
# Rodar o Gradient Descent 3 vezes
n = 100
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=n)
mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.001, precision=0.0001, max_iter=n)
high_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.002, precision=0.0001, max_iter=n)

# Experiment
insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9, multiplier=0.25, precision=0.0001, max_iter=n)

# Projetando a redução de custo para cada iteração
plt.figure(figsize=[18,6])

plt.title('Efeito da taxa de aprendizado')
plt.xlim([0, n])
plt.ylim(0, 50)
plt.xlabel('Número de iterações', fontsize=16)
plt.ylabel('Custo', fontsize=16)

# Valores para nosso gráfico
# 1 - Dados eixo Y: Converter a lista para um array numpy
low_values = np.array(low_gamma[1])

# 2 - Dados eixo X: Criar uma lista de 0 até n + 1
iteration_list = list(range(0, n+1))

# Projetando a taxa de aprendizado menor
plt.plot(iteration_list, g(low_values), color='lightgreen', linewidth=3)
plt.scatter(iteration_list, g(low_values), color='lightgreen', s=80)

# Projetando a taxa de aprendizado médio
plt.plot(iteration_list, g(np.array(mid_gamma[1])), color='steelblue', linewidth=3)
plt.scatter(iteration_list, g(np.array(mid_gamma[1])), color='steelblue', s=80)

# Projetando a taxa de aprendizado maior
plt.plot(iteration_list, g(np.array(high_gamma[1])), color='b', linewidth=3)
plt.scatter(iteration_list, g(np.array(high_gamma[1])), color='b', s=80)

# Projetando a taxa de aprendizado insana
plt.plot(iteration_list, g(np.array(insane_gamma[1])), color='r', linewidth=3)
plt.scatter(iteration_list, g(np.array(insane_gamma[1])), color='r', s=80)
plt.grid()
```
