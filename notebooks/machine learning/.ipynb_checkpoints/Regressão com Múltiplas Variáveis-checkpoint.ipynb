### Simplificação do Modelo e BIC(Bayesian information criterion)

```python
# Modelo original com preço em log e todos os features
X_incl_const = sm.add_constant(X_train.values)
X_incl_const_df = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const_df)
results = model.fit()
org_coef = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})

print(org_coef)
print(results.bic)
print(results.rsquared)
```

```python
# Modelo reduzido #1 excluindo INDUS
X_incl_const = sm.add_constant(X_train.values)
X_incl_const_df = sm.add_constant(X_train)
X_incl_const_df = X_incl_const_df.drop(['INDUS'], axis=1)
model = sm.OLS(y_train, X_incl_const_df)
results = model.fit()
coef_minus_indus = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})

print(coef_minus_indus)
print(results.bic)
print(results.rsquared)
```

```python
# Modelo reduzido #2 excluindo INDUS e AGE
# Quanto menor o valor de BIC, melhor
X_incl_const = sm.add_constant(X_train.values)
X_incl_const_df = sm.add_constant(X_train)
X_incl_const_df = X_incl_const_df.drop(['INDUS', 'AGE'], axis=1)
model = sm.OLS(y_train, X_incl_const_df)
results = model.fit()
reduced_coef = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})

print(reduced_coef)
print(results.bic)
print(results.rsquared)
```

```python
frames = [org_coef, coef_minus_indus, reduced_coef]
pd.concat(frames, axis=1, sort=True)
```

### Residuals e Gráficos de Residuals

```python
# Modelo modificado: Transformado (utlizando preços em log) e simplificado (dropamos duas features 'INDUS' e 'AGE')
prices = np.log(data['price']) # Utilizando preços em log
features = data.drop(['price','INDUS','AGE'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, 
                                                    test_size=0.2, 
                                                    random_state=10)

# Utilizando Statsmodel
X_incl_const = sm.add_constant(X_train.values)
X_incl_const_df = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const_df)
results = model.fit()
```

```python
# Residuals
residuals = y_train - results.fittedvalues
```

```python
residuals.describe()
```

```python
results.resid
```

```python
# Gráfico dos Preços Verdadeiros vs Preços Previstos
corr = round(y_train.corr(results.fittedvalues), 2)
plt.figure(figsize=(10,8))
plt.scatter(x=y_train, y=results.fittedvalues, c='navy', alpha=0.8)
plt.plot(y_train, y_train, color='cyan')
plt.xlabel('Preços verdadeiros (log) $y _i$', fontsize=15)
plt.ylabel('Preços previstos (log) $\hat y_i$', fontsize=15)
plt.title(f'Preços Verdadeiros vs Preços Previstos (log): $y _i$ vs $\hat y_i$ (Corr {corr})')

# Mean Squared Error & R-Squared
reduced_log_mse = round(results.mse_resid, 3)
reduced_log_rsquared = round(results.rsquared, 3)

print(reduced_log_mse, reduced_log_rsquared)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=np.e**y_train, y=np.e**results.fittedvalues, c='green', alpha=0.8)
plt.plot(np.e**y_train, np.e**y_train, color='blue')
plt.xlabel('Preços verdadeiros $y _i$', fontsize=15)
plt.ylabel('Preços previstos $\hat y_i$', fontsize=15)
plt.title(f'Preços Verdadeiros vs Preços Previstos: $y _i$ vs $\hat y_i$ (Corr {corr})')
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=results.fittedvalues, y=results.resid, c='navy', alpha=0.7)
plt.xlabel('Preços previstos (log) $\hat y _i$', fontsize=15)
plt.ylabel('Residuals', fontsize=15)
plt.title(f'Residuals vs Valores Ajustados', fontsize=16)
```

```python
# Distribuição dos Residuals (preço em log) - Checando por normalidade
resid_mean = round(results.resid.mean(), 3)
resid_mean
```

```python
resid_skew = round(results.resid.skew(), 3)
resid_skew
```

```python
plt.figure(figsize=(10,8))
sns.distplot(results.resid, color='navy')
plt.title(f'Modelo do preço em log: residuals Skew ({resid_skew}) Mean ({resid_mean})')
```

```python
# Modelo original: preços normal e todos as features
prices = data['price'] 
features = data.drop(['price'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, 
                                                    test_size=0.2, 
                                                    random_state=10)

# Utilizando Statsmodel
X_incl_const = sm.add_constant(X_train.values)
X_incl_const_df = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const_df)
results = model.fit()

# Gráfico dos Preços Verdadeiros vs Preços Previstos
corr = round(y_train.corr(results.fittedvalues), 2)
plt.figure(figsize=(10,8))
plt.scatter(x=y_train, y=results.fittedvalues, c='indigo', alpha=0.8)
plt.plot(y_train, y_train, color='cyan')
plt.xlabel('Preços verdadeiros $y _i$', fontsize=15)
plt.ylabel('Preços previstos $\hat y_i$', fontsize=15)
plt.title(f'Preços Verdadeiros vs Preços Previstos: $y _i$ vs $\hat y_i$ (Corr {corr})')

# Mean Squared Error & R-Squared
full_normal_mse = round(results.mse_resid, 3)
full_normal_rsquared = round(results.rsquared, 3)

print(full_normal_mse, full_normal_rsquared)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=results.fittedvalues, y=results.resid, c='indigo', alpha=0.7)
plt.xlabel('Preços previstos $\hat y _i$', fontsize=15)
plt.ylabel('Residuals', fontsize=15)
plt.title(f'Residuals vs Valores Ajustados', fontsize=16)
```

```python
resid_mean = round(results.resid.mean(), 3)
resid_skew = round(results.resid.skew(), 3)
plt.figure(figsize=(10,8))
sns.distplot(results.resid, color='indigo')
plt.title(f'Modelo do preço: residuals Skew ({resid_skew}) Mean ({resid_mean})')
```

```python
# Modelo omitindo features chaves (usando preço em log)
prices = np.log(data['price']) # Utilizando preços em log
features = data.drop(['price', 'INDUS', 'AGE', 'LSTAT', 'RM', 'NOX', 'CRIM'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, 
                                                    test_size=0.2, 
                                                    random_state=10)

# Utilizando Statsmodel
X_incl_const = sm.add_constant(X_train.values)
X_incl_const_df = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const_df)
results = model.fit()

# Gráfico dos Preços Verdadeiros vs Preços Previstos
corr = round(y_train.corr(results.fittedvalues), 2)
plt.figure(figsize=(10,8))
plt.scatter(x=y_train, y=results.fittedvalues, c='#e74c3c', alpha=0.8)
plt.plot(y_train, y_train, color='cyan')
plt.xlabel('Preços verdadeiros (log) $y _i$', fontsize=15)
plt.ylabel('Preços previstos (log) $\hat y_i$', fontsize=15)
plt.title(f'Preços Verdadeiros vs Preços Previstos com variáveis omitidas: $y _i$ vs $\hat y_i$ (Corr {corr})')

# Mean Squared Error & R-Squared
omitted_var_mse = round(results.mse_resid, 3)
omitted_var_rsquared = round(results.rsquared, 3)

print(ommited_var_mse, ommited_var_rsquared)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=results.fittedvalues, y=results.resid, c='#e74c3c', alpha=0.7)
plt.xlabel('Preços previstos $\hat y _i$', fontsize=15)
plt.ylabel('Residuals', fontsize=15)
plt.title(f'Residuals vs Valores Ajustados', fontsize=16)
```

```python
pd.DataFrame({'R-Squared': [reduced_log_rsquared, full_normal_rsquared, omitted_var_rsquared], 
              'MSE': [reduced_log_mse, full_normal_mse, omitted_var_mse],
             'RMSE': np.sqrt([reduced_log_mse, full_normal_mse, omitted_var_mse])}, 
             index=['Reduced Log Model', 'Full Normal Price Model', 'Omitted Variable Model'])
```

```python
print(f'1 Desvio Padrão dos preços (log) é {np.sqrt(reduced_log_mse)}')
```

```python
print(f'2 Desvio Padrão dos preços (log) é {2*np.sqrt(reduced_log_mse)}')
```

```python
upper_bound = np.log(30) + 2 * np.sqrt(reduced_log_mse)
print(f'The upper bound in log prices for a 95% prediction internval is {upper_bound}')
print(f'The upper bound in normal prices is ${np.e**upper_bound * 1000}')

lower_bound = np.log(30) - 2 * np.sqrt(reduced_log_mse)
print(f'The lower bound in log prices for a 95% prediction internval is {lower_bound}')
print(f'The lower bound in normal prices is ${np.e**lower_bound * 1000}')
```
