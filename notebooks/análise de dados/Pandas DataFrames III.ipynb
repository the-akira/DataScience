### Setando Múltiplos Valores em um DataFrame

```python
bond = pd.read_csv('https://dadosdatascience.netlify.com/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head()
```

```python
bond.loc['Dr. No']
```

```python
mask = bond['Actor'] == 'Sean Connery'
```

```python
bond.loc[mask, 'Actor'] = 'Sir Sean Connery'
```

```python
bond.head()
```

### Setando um Novo Valor para uma Célula ou Linha Específica

```python
bond = pd.read_csv('https://dadosdatascience.netlify.com/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head()
```

```python
bond.loc['Dr. No', 'Actor'] = 'Sir Sean Connery'
```

```python
bond.loc['Dr. No']
```

```python
bond.loc['Dr. No', ['Box Office', 'Budget', 'Bond Actor Salary']] = [448800000, 7000000, 600000]
```

```python
bond.loc['Dr. No']
```

### Renomear Rótulos do Índice ou Colunas em um DataFrame

```python
bond = pd.read_csv('https://dadosdatascience.netlify.com/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head()
```

```python
bond.rename(columns = {'Year' : 'Release Date', 'Box Office' : 'Revenue'}, inplace=True)
```

```python
bond.head()
```

```python
bond.rename(index = {
                    'Dr. No' : 'Doctor No', 
                    'GoldenEye' : 'Golden Eye',
                    'The World Is Not Enough': 'The Best'
                    }, inplace=True)
```
