## Rankeando Valores com o Método rank()

```python
nba = pd.read_csv('https://dadosdatascience.netlify.com/nba.csv').dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba.head()
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending=False).astype('int')
```

```python
nba.head()
```

```python
nba.sort_values(by='Salary', ascending=False).head()
```
