## Filtrando DataFrames

```python
import pandas as pd
```

```python
df = pd.read_csv('https://dadosdatascience.netlify.com/employees.csv')
# df = pd.read_csv('https://dadosdatascience.netlify.com/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
```

```python
df[df['Salary'] > 110_000].head()
```

### Filtrando com mais de uma Condição

```python
df = pd.read_csv('https://dadosdatascience.netlify.com/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.head()
```

```python
mask1 = df['Gender'] == 'Male'
mask2 = df['Team'] == 'Marketing'

df[mask1 & mask2].head()
```

```python
mask_1 = df['Team'] == 'Legal'
mask_2 = df['Team'] == 'Sales'
mask_3 = df['Team'] == 'Product'

df[mask_1 | mask_2 | mask_3].head()
```

```python
mask_4 = df['Team'].isin(['Legal', 'Sales', 'Product', 'Marketing'])
df[mask_4].head()
```

### O Método between()

```python
df = pd.read_csv('https://dadosdatascience.netlify.com/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.head()
```

```python
df[df['Salary'].between(60_000, 70_000)].head()
```

```python
df[df['Bonus %'].between(2.0, 5.0)].head()
```

```python
df[df['Start Date'].between('1991-01-01', '1992-01-01')].head()
```

```python
df[df['Last Login Time'].between('08:30AM', '12:00PM')].head(8)
```

### O Método duplicated()

```python
df = pd.read_csv('https://dadosdatascience.netlify.com/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.sort_values('First Name', inplace=True)
df.head()
```

```python
df[df['First Name'].duplicated()].head()
```

```python
df[df['First Name'].duplicated(keep='last')].head()
```

```python
df[~df['First Name'].duplicated(keep=False)]
```

### O Método drop_duplicates()

```python
df = pd.read_csv('https://dadosdatascience.netlify.com/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.head()
```

```python
len(df) # Quantidade de linhas
```

```python
df.drop_duplicates(subset=['First Name'], keep='last').head()
```

```python
df.drop_duplicates(subset=['First Name', 'Team']).head()
```
