### Exercício: Gerando uma Distribuição Normal com **média** = 15 e **desvio padrão** = 4.3

```python
mean = 15
std = 4.3

nums = np.random.randn(1000)
nums = nums - np.mean(nums)
nums = nums/np.std(nums)
nums = nums * std + mean

plt.figure(figsize=(13,8))
plt.subplot(121)
plt.plot(nums, 's')
plt.subplot(122)
plt.hist(nums)

print(f'Essa distribuição tem uma média de {np.mean(nums)} e um desvio padrão de {np.std(nums)}')
```

```python
def angle_convert_plot():
    # Obtendo os inputs do usuário
    angle = eval(input('Qual ângulo para converter? '))
    unit = input('Qual unidade você entrou (radians or degrees)? ')
    
    if unit[0].lower() == 'r':
        rad = angle
        deg = np.rad2deg(angle)
    elif unit[0].lower() == 'd':
        deg = angle
        rad = np.deg2rad(angle)
    else:
        raise ValueError('Unidade desconhecida')
    
    # Converter para < 360
    deg = deg % 360
    rad = rad % (2 * np.pi)
    
    # Criando o gráfico
    plt.figure(figsize=(12,8))
    plt.plot([0,1],[0,0],'r',linewidth=2)
    plt.plot([0,np.cos(rad)],[0,np.sin(rad)],'r',linewidth=3)
    plt.axis('square')
    plt.axis([-1,1,-1,1])
    plt.grid()
    plt.show()
```

```python
angle_convert_plot()
```

### Teorema de Pythagoras

```python
a = 8
b = 10
c = int(np.sqrt(a**2 + b**2))
```

```python
# Desenhando o Triângulo
plt.figure(figsize=(10,8))
plt.plot([0,a],[0,0],'k',linewidth=2)
plt.plot([0,0],[0,b],'k')
plt.plot([0,a],[b,0],'k')

plt.plot([.3,.3],[0,.3],'k',linewidth=1)
plt.plot([0,.3],[.3,.3],'k')

plt.text(a/3, -0.35, 'a=' + str(a), fontsize=15)
plt.text(0.2, b/3, 'b=' + str(b), fontsize=15)
plt.text(a/2+0.1, b/2, 'c=' + str(c), fontsize=15)

plt.axis('square')
axlim = np.max((a,b)) + .5
plt.axis([-.5,axlim,-.5,axlim])
plt.axis('off')
```

```python
z = np.complex(3,4)

# Computando a distância da origem
magnitude_1 = np.sqrt(np.real(z)**2 + np.imag(z)**2)
magnitude_2 = np.abs(z)

print(magnitude_1,magnitude_2)
```

```python
# Extraindo os ângulos
angulo_1 = np.arctan2(np.imag(z), np.real(z))
angulo_2 = np.angle(z)

print(angulo_1, angulo_2)
```

```python
thsym = sym.pi*7/5
theta = float(thsym)

# Projetando o Círculo Unidade
x = np.linspace(0, 2*np.pi, 200)
plt.figure(figsize=(10,10))
plt.plot(np.cos(x), np.sin(x), color='gray')
plt.plot([-1.1,1.1], [0,0], '--', color='gray')
plt.plot([0,0], [-1.1,1.1], '--', color='gray')

# Projetando o arco
plt.plot(np.cos(np.linspace(0, theta)), np.sin(np.linspace(0, theta)), 'k', linewidth=3)
plt.plot([0,np.cos(theta)],[0,np.sin(theta)], 'k:')
plt.plot(np.cos(theta),np.sin(theta),'ko')

# Projetando os componentes separados (seno e cosseno)
plt.plot([0,0],[0,np.sin(theta)],'r')
plt.plot([0,np.cos(theta)],[np.sin(theta),np.sin(theta)],'r:')

plt.plot([0,np.cos(theta)],[0,0],'b')
plt.plot([np.cos(theta), np.cos(theta)], [0, np.sin(theta)], 'b:')

plt.title(f'$\\theta = {sym.latex(thsym)}$')
plt.show()
```

### Fórmula de Euler

```python
k = np.pi/4
m = 2.3

# Notação Euler
euler = m*np.exp(1j*k)
# Notação Cosseno + Seno
cis = m*(np.cos(k) + 1j * np.sin(k))

print(euler)
print(cis)
```

```python
magnitude = np.abs(euler)
angulo = np.angle(euler)

print(magnitude)
print(angulo)
```

```python
plt.figure(figsize=(10,10))
plt.polar([0,angulo],[0,magnitude],'b')
plt.polar(k, m, 'ko')
plt.show()
```

```python
def euler_from_cos_sin():
    re = eval(input('Parte Cosseno: '))
    im = eval(input('Parte Seno: '))
    
    m = np.sqrt(re**2 + im**2)
    print(m)
    k = np.arctan2(im, re)
    print(k)
    
    plt.figure(figsize=(10,10))
    plt.polar([0,k],[0,m],'k')
    plt.title('me$^{i\\phi}$, m=%g, $\\phi$=%g' %(m,k))
    plt.thetagrids([0,45,130,200,222])
    plt.show()
```

```python
euler_from_cos_sin()
```

### Explodindo Euler Aleatoriamente

```python
n_vetores = 200
m = np.random.rand(n_vetores)
k = np.random.rand(n_vetores) * 2 * np.pi

plt.figure(figsize=(10,10))

for i in range(0, n_vetores):
    r = np.random.rand()
    if r < 0.4:
        color = [1,.2,.7]
    elif r > 0.4 and r < 0.8:
        color = [.7,.2,1]
    else:
        color = [0,1,0]
    plt.polar([0,k[i]],[0,m[i]],color=color,linewidth=3.3)

plt.axis('off')
plt.show()
```

### Linhas Curvas Aleatórias

```python
t = np.linspace(0, 8*np.pi, 1000)

r1 = np.random.rand()
r2 = np.random.rand()

x = np.cos(r1*t)
y = np.sin(r2*t)

plt.figure(figsize=(10,10))
plt.plot(x, y,'k')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.axis('off')
plt.title(f'r$_1$={np.round(r1, 2)}, r$_2$={np.round(r2, 2)}')
plt.show()
```

```python
t = np.linspace(0, 8*np.pi, 1000)

r1 = 1
r2 = 1

x = np.cos(r1*t)
y = np.sin(r2*t)

plt.figure(figsize=(10,10))
plt.plot(x, y,'k')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.axis('off')
plt.title(f'r$_1$={np.round(r1, 2)}, r$_2$={np.round(r2, 2)}')
plt.show()
```

```python
t = np.linspace(0, 8*np.pi, 1000)

r1 = np.random.rand()
r2 = np.random.rand()

x = np.cos(r1*t)
y = np.sin(r2*t)

x = np.append(x,x[0])
y = np.append(y,y[0])

plt.figure(figsize=(10,10))
plt.plot(x, y,'k')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.axis('off')
plt.title(f'r$_1$={np.round(r1, 2)}, r$_2$={np.round(r2, 2)}')
plt.show()
```
