# Gráficos e Visualizações

```python
import matplotlib.pyplot as plt 
from matplotlib.patches import Polygon
from IPython.display import display, Math
import sympy as sym
import numpy as np
```

### Projetando uma Coordenada (3,5)

```python
x = 3 
y = 5

plt.figure(figsize=(11,7))
plt.plot(x, y, 'ks')
plt.axis([-6,6,-6,6])
plt.grid()
```

```python
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]

plt.figure(figsize=(11,7))
plt.grid()

for i in range(len(x)):
    plt.plot(x[i], y[i], 'o', label=f'point {i}')

plt.legend()
```

```python
plt.figure(figsize=(11,7))
plt.grid()
plt.plot(4, 2.3, 'r+')

axis = plt.gca()
ylim = axis.get_ylim() # Obtendo limite y
xlim = axis.get_xlim() # Obtendo limite x
axis.set_ylim([ylim[0], 5.39]) # Alterando limite y
print(ylim, xlim)
```

```python
x = sym.symbols('x')
y = x**2 - 3*x

plt.figure(figsize=(11,7))
plt.xlabel('x')
plt.ylabel(f'f(x) = {y}')
plt.grid()

for point in range(-11,12):
    plt.plot(point, y.subs(x, point), 'o')
```

```python
plt.figure(figsize=(8,8))
plt.plot([0,2],[2,2],'k')
plt.plot([0,2],[0,0],'k')
plt.plot([0,0],[0,2],'k')
plt.plot([2,2],[0,2],'k')
plt.axis('square')
plt.axis([-3,5,-3,5])
plt.grid()
```

```python
# Representando y = mx + b

x = [-5,5]
m = 0.8
b = 2.2

y = [340, 2345]

for i in range(len(x)):
    y[i] = m * x[i] + b
    
print(x,y)
plt.figure(figsize=(8,8))
plt.plot(x, y, label=f'y={m}x+{b}', linewidth=2, color='r')
plt.axis('square')
plt.grid()
plt.legend()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_ylim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')
plt.show()
```

```python
x = [-5,5]
m = [0.7,-1.25]
b = [-2,0.75,]

plt.figure(figsize=(11,7))
plt.grid()

for i in range(len(m)):
    y = m[i] * np.array(x) + b[i]
    plt.plot(x, y, label=f'y = {m[i]}x + {b[i]}', linewidth=2)

plt.legend()
plt.xlim(x)
plt.ylim(x)
axis = plt.gca()
plt.plot(axis.get_ylim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')
plt.show()
```

### Gráficos de Funções Racionais

Exemplos de funções racionais:

$$ f(x) = \frac{x^3 + x^2 + 1}{-x^3 + x - 1} $$

$$ g(x) = \frac{1 + \sqrt{x}}{1} $$

```python
x = range(-3,4)
y = np.zeros(len(x))

for i in range(len(x)):
    y[i] = 2 - x[i]**2

plt.figure(figsize=(11,7))
plt.plot(x, y, 's-')
plt.grid()
```

```python
x = np.linspace(-3,3,100)
y1 = 2 + np.sqrt(abs(x))
y2 = 2 + abs(x)**(1/3)

plt.figure(figsize=(11,7))
plt.plot(x, y1, 'm-')
plt.plot(x, y2, 'k-')
plt.grid()
plt.show()
```

```python
x = np.linspace(-4,4,100)
e = range(-1,4)

plt.figure(figsize=(11,7))
plt.grid()

for i in e:
    y = x**i
    plt.plot(x, y, label=f'$y=x^{i}$', linewidth=2.8)

plt.legend()
plt.xlim([x[0],x[-1]])
plt.ylim([-20,20])
```

### Gráficos com SymPy

```python
x = sym.symbols('x')
y = x**2

p = sym.plotting.plot(y, show=False)
p.xlim = [0,50]
p[0].line_color = 'red'
p.title = 'Gráfico'
p.show()
```

```python
x, a = sym.symbols('x a')

expressao = a/x 

n = 1

p = sym.plotting.plot(expressao.subs(a,n),(x,-5,5), show=False)
p[0].label = f'$y={sym.latex(expressao.subs(a,n))}$'
p[0].line_color = 'red'

p.extend(sym.plotting.plot(expressao.subs(a, 2), show=False))
p[1].label = f'$y={sym.latex(expressao.subs(a,2))}$'

p.extend(sym.plotting.plot(expressao.subs(a, 5), show=False))
p[2].label = f'$y={sym.latex(expressao.subs(a,5))}$'
p[2].line_color = 'black'

p.ylim = [-5,5]
p.legend = True
p.show()
```

### Exercícios

```python
# Começamos definindo nossa função
y = a/(x**2 - a)

p = None
p = sym.plotting.plot(y.subs(a,1),(x,-5,5),show=False)
p[0].label = f'${sym.latex(y.subs(a,1))}$'

for i in range(2,5):
    p.extend(sym.plotting.plot(y.subs(a,i),(x,-5,5),show=False))
    p[i-1].label = f'${sym.latex(y.subs(a,i))}$'
    p[i-1].line_color = list(np.random.rand(3))

p.ylim = [-10,10]
p.legend = True
p.show()
```

### Criando Imagens através de Matrizes

```python
matriz = [[1,2],[1,4]]

plt.figure(figsize=(7,7))
plt.imshow(matriz)
plt.xticks([0,1])
plt.yticks([0,1])
plt.show()
```

```python
A = np.zeros((10,14))

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        A[i,j] = 3*i - 4*j

plt.figure(figsize=(7,7))
plt.imshow(A)
plt.plot([0,3],[8,2],'k',linewidth=3)

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        plt.text(j, i, int(A[i,j]), horizontalalignment='center', verticalalignment='center')

plt.set_cmap('rainbow')
```

### Exercício: Criando um tabuleiro de Xadrez

$$ c_{i,j} = -1^{i+j} $$

```python
# Soft-Coding N
N = 10
C = np.zeros((N,N))
display(Math(sym.latex(sym.sympify(C))))

for i in range(N):
    for j in range(N):
        C[i,j] = (-1)**(i+j)

display(Math(sym.latex(sym.sympify(C))))
plt.figure(figsize=(7,7))
plt.set_cmap('gray')
plt.tick_params(labelleft=False, labelbottom=False)
plt.imshow(C)
plt.show()
```

### Desenhando Polígonos

```python
# Gerando os pontos que irão compor o polígono
y = np.array([[1,1],[2,3],[3,1]])
p = Polygon(y, color='k', alpha=0.9)

y1 = np.array([[2,2],[2.5,4],[3.5,1]])
p1 = Polygon(y1, color='g', alpha=0.8)

fig, ax = plt.subplots()
fig.set_figheight(7)
fig.set_figwidth(10)
ax.add_patch(p)
ax.add_patch(p1)
ax.set_ylim([0,4])
ax.set_xlim([0,4])
plt.grid()
plt.show()
```

### Exercício

$ f(x) = -x^2 $

```python
x = np.linspace(-2,2,101)
f = -x**2

pontos = np.vstack((x,f)).T
p = Polygon(pontos, facecolor='g', alpha=0.4)

y = np.array([[-0.5,-4],[-0.5,-2.5],[0.5,-2.5],[0.5,-4]])
p1 = Polygon(y, facecolor='k')

fig, ax = plt.subplots()
fig.set_figheight(7)
fig.set_figwidth(10)
ax.add_patch(p)
ax.add_patch(p1)
plt.plot(x, f, 'k')
plt.plot([-2,2],[-4,-4],'k')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,7))
# Desenhando um tabuleiro de xadrez com números em cima
m = 8
n = 4

# Inicializando a matriz
C = np.zeros((m,4))

# Populando a matriz
for i in range(m):
    for j in range(n):
        C[i,j] = (-1)**(i+j)
    
# Apresentando os números
for i in range(m):
    for j in range(n):
        plt.text(j, i, i+j, horizontalalignment='center', verticalalignment='center', fontdict=dict(color='g'))

plt.imshow(C)
plt.set_cmap('gray')
plt.axis('off')
plt.show()
```
