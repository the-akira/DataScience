```python
x = sym.symbols('x')

# Definindo uma funÃ§Ã£o e sua derivada
f = x**2
df = sym.diff(f)

# O valor pelo qual vamos computar a linha tangente
xa = 0.6

# Obter a funÃ§Ã£o e o valor derivado em xa
fa = f.subs(x, xa)
dfa = df.subs(x, xa)

# Gerando valores e lambdifycando as funÃ§Ãµes
xx = np.linspace(-2, 2, 200)
f_func = sym.lambdify(x, f)(xx)
df_fun = sym.lambdify(x, df)(xx)

# Computar a linha tangente
linha_tan = dfa * (xx-xa) + fa

plt.plot(xx, f_func, label='f(x)', linewidth=3)
plt.plot(xx, linha_tan, label='tangente', linewidth=3)
plt.plot(xa, fa, 'ro')
plt.axis('square')
plt.axis([-2,2,-0.5,2])
ax = plt.gca()
plt.plot(ax.get_xlim(),[0,0],'k--')
plt.plot([0,0], ax.get_ylim(), 'k--')
plt.legend()
plt.grid()
plt.show()
```

```python
def computar_tangente(f, xa, bounds):
    # Definindo a derivada e valores no ponto xa
    df = sym.diff(f)
    fa = f.subs(x, xa)
    dfa = df.subs(x, xa)
    # Calcular a linha tangente
    xx = np.linspace(bounds[0], bounds[1], 200)
    return dfa * (xx - xa) + fa
```

```python
x = sym.symbols('x')
f = x**2
xx = np.linspace(-4, 4, 200)
f_func = sym.lambdify(x, f)(xx)

linha_tangente = computar_tangente(f, 1.7, xx[[0,-1]])

plt.axis([-5,5,0,15])
plt.plot(xx, f_func)
plt.plot(xx, linha_tangente)
plt.grid()
plt.show()
```

```python
a, x = sym.symbols('a, x')

expressao = x**2 * sym.exp(-a*x**2)

a_range = np.arange(0, 2.25, 0.25)
x_range = np.linspace(-3, 3, 100)

# Setando a figura
fig, ax = plt.subplots(1, 2)

for ai in a_range:
    fx = expressao.subs(a, ai)
    dfx = sym.diff(fx)
    pontos_criticos = sym.solve(dfx)
    
    # Gerando o grÃ¡fico
    ax[0].plot(x_range, sym.lambdify(x, fx)(x_range))
    ax[1].plot(x_range, sym.lambdify(x, dfx)(x_range))
    
    # Testando por pontos crÃ­ticos
    if 1 in pontos_criticos:
        display(Math('\\Rightarrow %s \\text{possui um ponto crÃ­tico em x=1}' %sym.latex(fx)))
    elif 2 in pontos_criticos:
        display(Math('\\Rightarrow %s \\text{possui um ponto crÃ­tico em x=2}' %sym.latex(fx)))
    else:
        display(Math('\\Rightarrow %s \\text{NÃ£o possui pontos crÃ­ticos em x=1 e x=2}' %sym.latex(fx)))

ax[0].set_ylim([0,2])
ax[0].set_title('FunÃ§Ã£o')
ax[0].plot([1,1],[0,2],'--',color='gray')
ax[0].plot([2,2],[0,2],'--',color='gray')
ax[1].set_ylim([-2,2])
ax[1].set_title('Derivada')
ax[1].plot([1,1],[-2,2],'--',color='gray')
ax[1].plot([2,2],[-2,2],'--',color='gray')
```

### Computando a Ã¡rea entre duas curvas 

```python
x = sym.symbols('x')

sym_f = x**2
sym_g = x

f = sym.lambdify(x, sym_f)
g = sym.lambdify(x, sym_g)

xx = np.linspace(-2, 2, 55)

# Adicionando patch
xpatch = np.linspace(0, 1, 10)
ypatch = np.vstack((g(xpatch), f(xpatch))).T

fig, ax = plt.subplots()
ax.add_patch(Polygon(ypatch, facecolor='k', alpha=0.8))

plt.plot(xx, f(xx), linewidth=3)
plt.plot(xx, g(xx), 'r', linewidth=3)
plt.legend(['$ f(x)=%s $' %sym.latex(sym_f), '$ g(x)=%s $' %sym.latex(sym_g)])
plt.axis([-0.25, 1.25, -0.5, 1.5])
```

### Computando a Ã¡rea entre duas funÃ§Ãµes

A Ã¡rea entre duas funÃ§Ãµes Ã© dada pela fÃ³rmula $A = \int_{a}^{b}[f(x)-g(x)]dx$

Em nosso exemplo, $f(x)=x^2$ e $g(x)=x$

Sendo assim, $A = \int_{a}^{b}(x-x^2)dx$

NÃ³s vamos computar a Ã¡rea entre dois pontos que se cruzam, ou seja, onde as duas funÃ§Ãµes sÃ£o iguais. Isto Ã© dado pelas duas soluÃ§Ãµes Ã  equaÃ§Ã£o.

$$x^2=x$$

As duas soluÃ§Ãµes sÃ£o $x=0$ e $x=1$

Isto nos fornece a Integral definida de $$A = \int_{0}^{1}(x-x^2)dx$$

```python
fg_intersect = sym.solve(sym_f - sym_g)
display(Math('f(x)\\text{ e } g(x) \\text{ intersecÃ§Ã£o em } x= %g \\text{ e } x = %g' %(fg_intersect[0], fg_intersect[1])))
```

```python
# Computar a Ã¡rea
A = sym.integrate(sym_g - sym_f, (x, fg_intersect[0], fg_intersect[1]))
display(Math('\\text {A Ã¡rea entre as duas funÃ§Ãµes Ã© }%s' %A))
```

```python
# Computar a Ã¡rea entre duas curvas
x = sym.symbols('x')
f1_sym = sym.cos(x)
f2_sym = x

xx = np.linspace(0, np.pi/3, 100)
f1 = np.cos(xx)
f2 = xx

f_intersect = np.argmin(abs(f1-f2))

# Calculando a Ã¡rea
A = sym.integrate(f1_sym - f2_sym, (x, xx[0], xx[f_intersect]))

trace_X = np.concatenate((xx[0:f_intersect],xx[f_intersect:0:-1]))
trace_Y = np.concatenate((f1[0:f_intersect],f2[f_intersect:0:-1]))

pontos = np.vstack((trace_X, trace_Y)).T
p = Polygon(pontos, facecolor='k', alpha=0.6)

fig, ax = plt.subplots()
ax.add_patch(p)

plt.title(f'A Ã¡rea sombreada Ã© {A}')
plt.plot(xx, f1, xx, f2, linewidth=2)
plt.show()
```
