### O Método agg()

```python
fortune = pd.read_csv('https://dadosdatascience.netlify.com/fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head(10)
```

```python
sectors.agg({'Revenue': 'sum', 'Profits': 'sum', 'Employees': 'mean'})
```

```python
sectors.agg(['size', 'sum', 'mean'])
```

### Iterando através dos Grupos

```python
fortune = pd.read_csv('https://dadosdatascience.netlify.com/fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head(10)
```

```python
df = pd.DataFrame(columns=fortune.columns)
```

```python
df
```

```python
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, 'Revenue')
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

```python
cities = fortune.groupby('Location')
dados = pd.DataFrame(columns=fortune.columns)
dados
```

```python
for city, data in cities:
    highest_revenue_in_city = data.nlargest(1, 'Revenue')
    dados = dados.append(highest_revenue_in_city)
```
