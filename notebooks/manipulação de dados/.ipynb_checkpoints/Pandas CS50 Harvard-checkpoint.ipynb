## Explorando um Conjunto de Dados

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv('https://dadosdatascience.netlify.com/TemperaturasGlobais.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.iloc[:,:2]
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['LandAverageTemperature'])
plt.title('Average Land Temperature 1750-2015')
plt.xlabel('Year')
plt.ylabel('Average Land Temperature')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.scatter(df['LandAverageTemperature'].index, y = df['LandAverageTemperature'])
plt.title('Average Land Temperature 1750-2015')
plt.xlabel('Year')
plt.ylabel('Average Land Temperature')
plt.show()
```

```python
print(type(df['dt'][0]))
```

```python
times = pd.DatetimeIndex(df['dt'])
print(times)
grouped = df.groupby([times.year]).mean()
```

```python
plt.figure(figsize=(15,5))
plt.plot(grouped['LandAverageTemperature'])
plt.title('Yearly Average Land Temperature 1750-2015')
plt.xlabel('Year')
plt.ylabel('Yearly Averrage Land Temperature')
plt.show()
```

```python
grouped.head()
```

```python
df[times.year == 1752]
```

```python
df[np.isnan(df['LandAverageTemperature'])]
```

```python
df['LandAverageTemperature'] = df['LandAverageTemperature'].fillna(method='ffill')
```

```python
grouped = df.groupby([times.year]).mean()

plt.figure(figsize = (15, 5))
plt.plot(grouped['LandAverageTemperature'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression as LinReg
```

```python
x = grouped.index.values.reshape(-1,1)
y = grouped['LandAverageTemperature'].values
```

```python
reg = LinReg()
reg.fit(x, y)
y_preds = reg.predict(x)
print('Acurácia: ' + str(reg.score(x, y)))
```

```python
plt.figure(figsize = (15,5))
plt.title("Linear Regression")
plt.scatter(x = x, y = y_preds)
plt.scatter(x = x, y = y, c = 'r')
```
