# Pandas

## Alterando Opções

Começamos importando as bibliotecas necessárias

```python
import pandas as pd 
import numpy as np
```

Vamos gerar uma matriz de dados de números aleatórios entre 0 e 99 de 1000 linhas e 50 colunas 

```python
data = np.random.randint(0, 100, [1000, 50])
```

Visualizamos os dados

```python
data
```

Vamos agora converter os dados em um DataFrame

```python
df = pd.DataFrame(data)
```

Verificamos o formato de nosso DataFrame e confirmamos que ele possui respectivamente 1000 linhas e 50 colunas

```python
df.shape
```

Visualizamos as 12 primeiras entradas

```python
df.head(12)
```

Alteramos o padrão de máximo de linhas exibidas para 15

```python
pd.options.display.max_rows = 15
```

```python
df
```

Alteramos o padrão de máximo de colunas exibidas para 15

```python
pd.options.display.max_columns = 15
```

```python
df
```

Obtemos o valor atual do atributo max_rows

```python
pd.get_option('max_rows')
```

Obtemos o valor atual do atributo max_columns

```python
pd.get_option('max_columns')
```

Alteramos o valor de max_rows para 13

```python
pd.set_option('max_rows', 13)
```

```python
df
```

A opção precision nos permite alterar as casas decimais de precisão de nossos números

```python
dados = pd.DataFrame(np.random.randn(5, 5))
```

```python
dados
```

Vejamos nossa precisão atual

```python
pd.get_option('precision')
```

Alteramos nossa precisão para 10

```python
pd.set_option('precision', 10)
```

```python
dados
```
